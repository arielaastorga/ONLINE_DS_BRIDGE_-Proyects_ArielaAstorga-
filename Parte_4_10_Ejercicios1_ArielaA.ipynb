## Ejercicio 1

1. Crea dos variables numericas: un `int` y un `float`
2. Comprueba sus tipos
3. Sumalas en otra nueva
4. ¿De qué tipo es la nueva variable?
5. Elimina las dos primeras variables creadas

```python
# 1 Crea dos variables numericas: un `int` y un `float`
var_numero = 15
var_float = 16.5

# 2. Comprueba sus tipos
print(type(var_numero))
print(type(var_float))

# 3. Sumalas en otra nueva
suma_var = var_numero + var_float
print("La suma de las variables es",suma_var)

#4. ¿De qué tipo es la nueva variable?
print("La nueva variable es de tipo",type(suma_var))

# 5. Elimina las dos primeras variables creadas
del var_numero, var_float

 #print(var_numero, var_float)
```

## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

radianes = grados*(pi/180)

```python
grad = input("Ingresa el numero en grados, a convertir a radianes")
import math
phi = math.pi
grad2 = float(grad)
rad1 = grad2*(phi/180)
print("El resultado de la conversion es", rad1,"radianes")
```

## Ejercicio 3
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`

```python
print("Vamos a calcular el area de un paralelogramo")
base = input("Ingresa el valor de la base")
base_float = float(base)
altura = input("Ingresa el valor de la altura")
altura_float = float(altura)
Area = base_float * altura_float
print("El area del paralelogramo es", Area)
```

## Ejercicio 4
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`. En este caso debe ser capaz de admitir valores reales, con decimales, para base y altura en español. (PISTA: Recuerda que tendrás que hacer uso del método replace)

```python
print("Vamos a calcular el area de un paralelogramo")
base_esp = input("Ingresa el valor de la base")
base_gen = base_esp.replace(",",".") 
base_gfloat = float(base_gen)

altura_esp = input("Ingresa el valor de la altura")
altura_gen = altura_esp.replace(",",".") 
altura_gfloat = float(altura_gen)
Area2 = base_gfloat * altura_gfloat
print("El area del paralelogramo es", Area2)
```
