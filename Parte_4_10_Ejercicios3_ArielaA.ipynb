## Ejercicio 1
Para este ejercicio vamos a poner en práctica [las funciones built in](https://docs.python.org/3/library/functions.html).

1. Calcula el máximo de la lista: [4, 6, 8, -1]
2. Suma todos los elementos de la lista anterior
3. Redondea este float a 3 dígitos decimales: 63.451256965
4. Valor absoluto de: -74

```python
#1 1. Calcula el máximo de la lista: [4, 6, 8, -1]
lista1 =  [4, 6, 8, -1]
max1 = max(lista1)
print("El máximo de la lista anterior es", max1)

# 2. Suma todos los elementos de la lista anterior
suma_lista = sum(lista1)
print("La suma de la lista anterio es", suma_lista)

# 3. Redondea este float a 3 dígitos decimales: 63.451256965
num = 63.451256965
red_num = round(num,3)
print("El numero redondeado a tres decimales es",red_num)

# 4. Valor absoluto de: -74
num2 = -74
print("El valor absoluto de -74 es", abs(num2))

```

## Ejercicio 2
Para el siguiente string se pide imprimir por pantalla los siguientes casos

    "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."

1. Pásalo todo a mayúsculas
2. Pásalo todo a minúsculas
3. Solo la primera letra de cada palabra en mayúscula, el resto en minúscula
4. Crea una lista dividiéndolo por sus espacios
5. Sustituye las comas `,` por puntos y comas `;`
6. Elimina las `a` minúsculas

```python
# 1.Pásalo todo a mayúsculas
texto1 = ("En un lugar de la Mancha, de cuyo nombre no quiero acordarme.")
print ("Texto  en mayúsculas")
print ("\t",texto1.upper())

#2. Pásalo todo a minúsculas
print ("Texto  en minusculas")
print ("\t",texto1.lower())

#3. Solo la primera letra de cada palabra en mayúscula, el resto en minúscula
texto2 = (texto1.lower())
texto3 = texto2.split()

print("Texto primera letra de cada palabra en mayúscula, el resto en minúscula")
print("\t",'"',texto3[0].capitalize()+" "+texto3[1].capitalize()+" "+texto3[2].capitalize()+" "+texto3[3].capitalize()+" "+texto3[4].capitalize()+" "+texto3[5].capitalize()+" "+texto3[6].capitalize()+" "+texto3[7].capitalize()+" "+texto3[8].capitalize()+" "+texto3[9].capitalize()+" "+texto3[10].capitalize()+" "+texto3[11].capitalize(),'"')

#4. Crea una lista dividiéndolo por sus espacios

lista1 = texto1.split()
print("Frase como lista")  
print("\t",lista1)

#5. Sustituye las comas , por puntos y comas ;

coma_pos = lista1.index("Mancha,")  #Encuentro la posicion de la coma
coma_pos2 = coma_pos

elem = lista1[coma_pos2]  

nueva_pal = elem.replace(",",";") #Reemplazo por ;

lista1[coma_pos2]= nueva_pal

print("Sustituyo las comas por punto y coma")  
print("\t",lista1) #Imprimo nueva lista corregida

#6. Elimina las a minúsculas
texto2 = texto1.replace('a',"")
print("Elimino las a minúsculas") #Imprimo nueva lista corregida
print("\t",texto2)





```

## Ejercicio 3
1. Crea una lista con 3 elementos numéricos
2. Añade un cuarto elemento
3. Calcula la suma de todos
4. Elimina el segundo elemento de la lista
5. Añade otro elemento en la posicion 3 de la lista
6. Crea otra lista con 4 elementos y concaténala a la que ya tenías.
7. Ordena la lista de menor a mayor.
8. BONUS: Ordena la lista de mayor a menor.  
(*PISTA: Para hacer 8, tendrás que buscar en la información del método...*)  
  
  
NOTA: Pinta por pantalla el resultado de cada paso.

```python
#1. Crea una lista con 3 elementos numéricos
lista2 = [12, 13.5, 20]
print("1. Crea una lista con 3 elementos numéricos")
print(lista2)

#2. Añade un cuarto elemento
nuevo_elem = [18] 
lista3 = lista2 + nuevo_elem
print("2. Añade un cuarto elemento")
print(lista3)

#Otra forma
# lista2.insert(3,18)
# print("Otra forma", lista2)


#3. Calcula la suma de todos
suma_lista = sum (lista3)
print("3. Suma elementos nueva lista")
print(suma_lista)

#Elimina el segundo elemento de la lista
lista3.pop(1)
print("4. Elimina el segundo elemento de la lista")
print(lista3)

#Añade otro elemento en la posicion 3 de la lista
lista3.insert(2,22)
print("5. Añade otro elemento en la posicion 3 de la lista")
print(lista3)

#Crea otra lista con 4 elementos y concaténala a la que ya tenías.
lista4 = [45, 3.3, 66.6, 18]
lista3.extend(lista4)
print("6. Crea otra lista con 4 elementos y concaténala a la que ya tenías.")
print(lista3)

#Ordena la lista de menor a mayor.
print("7. Ordena la lista de menor a mayor.")
lista3.sort()
print(lista3)


#BONUS: Ordena la lista de mayor a menor.
print("8. BONUS: Ordena la lista de mayor a menor.")
lista4 = sorted(lista3, reverse=True)
print(lista4) 

```
