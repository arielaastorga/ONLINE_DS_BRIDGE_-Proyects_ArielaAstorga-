## Ejercicio 2
Haz un programa que detecte todos los duplicados de un elemento en una lista:
* Utiliza una variable duplicado para asignarle el valor del elemento del que queremos encontrar todos sus duplicados.
* El programa debe devolver todos los índices de los elementos duplicados.
* Aplícalo para encontrar los duplicados de "un", "es" y "binario" de:  

>```ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]```


`````python
# Funcion para encontrar elementos duplicados

def duplicados(lista):   #Funcion para encontrar duplicados

    recorrer = set()
    duplicados = set()
    indice = set()

    for item in lista:
        if item in recorrer:
            duplicados.add(item)
            indice.add(lista.index(item))
        else:
            recorrer.add(item)
    print("Los elementos duplicados son", list(duplicados))
    print("Los indices de los elementos duplicados son", list(indice)) 
    

# Aplicando la funcion en la lista de ejemplo

ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]

duplicados(ejer_3)    
`````
